```python
# Install gTTS if not already installed
!pip install gTTS

from gtts import gTTS
from IPython.display import Audio
import os

def robo_speak(text):
    tts = gTTS(text=text, lang='en')
    tts.save("voice.mp3")
    return Audio("voice.mp3", autoplay=True)

print("Welcome to RoboSpeaker 1.1 (Colab Edition)")

while True:
    x = input("Enter what you want me to speak (type 'q' to quit): ")
    if x.lower() == "q":
        robo_speak("bye bye friend")
        break
    else:
        display(robo_speak(x))
```
